```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.decomposition import TruncatedSVD, PCA
```

```python
planets = sns.load_dataset('planets')
planets.shape
```

# Data Aggregation

```python
planets.head()
```

```python
planets['mass'].sum()
```

```python
planets['mass'].mean()
```

```python
  planets['mass'].count()
```

```python
planets.drop(columns = 'method', inplace=True)
```

```python
planets.fillna(method="ffill", inplace=True)
planets
```

# SVD

```python
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=5, n_iter=7, random_state=42)
planets_svd = svd.fit_transform(planets)
planets_svd
```

# PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5, random_state=42)
planets_pca = pca.fit_transform(planets)
planets_pca
```
